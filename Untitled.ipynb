```python
from Bio import Entrez
from Bio import SeqIO
import pandas as pd
import os

# os.chdir(r"/mnt/c/Users/yschoi/sero_down")
Entrez.email = 'seok1213neo@snu.ac.kr'

serotypes = pd.read_excel(r"C:\Users\yschoi\Desktop\serotype\Vco_3.xlsx")
accessions = pd.DataFrame(serotypes, columns = ['ACC'])

acc_list = []

for number in list(range(306)):
    for accession in accessions.loc[number]:
        if "NZ" in accession:
            acc_list.append(accession)

id_list = []
            
for acc in acc_list:  #summary에나온 아이디로 검색하니까 나오더라!
    handle = Entrez.esummary(db="nucleotide", id= acc, retmode="xml")
    records = Entrez.parse(handle)

    for record in records:
        id_list.append(record['Id'])

ass_list = []
        
for id in id_list: #summary id로 검색해 나온 결과에서 assembly id 빼기
    handle = Entrez.efetch(db = 'nucleotide', id = id, rettype = 'gb', retmode = 'text')
    for record in SeqIO.parse(handle, 'genbank'):
        assembly = record.dbxrefs[2]
        assembly = assembly.split(":")[1]
        ass_list.append(assembly)
    
for ass in ass_list:
    number = ass_list.index(ass)
    os.system ('/bin/bash -c "esearch -db assembly -query {1} | elink -target nucleotide -name assembly_nuccore_insdc | efetch -format gb > {0}_{1}.gb"').format(acc_list[number], ass)
```
